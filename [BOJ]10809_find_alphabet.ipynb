# 알파벳 찾기
- source : [백준 10809](https://www.acmicpc.net/problem/10809)
- 난이도 : 브론즈 II
- 분류 : 구현, 문자열

## 문제
- 알파벳 소문자로만 이루어진 단어 S가 주어진다. 각각의 알파벳에 대해서, 단어에 포함되어 있는 경우에는 처음 등장하는 위치를, 포함되어 있지 않은 경우에는 -1을 출력하는 프로그램을 작성하시오. 
    
## 입력
- 첫째 줄에 단어 S가 주어진다. 단어의 길이는 100을 넘지 않으며, 알파벳 소문자로만 이루어져 있다.

## 출력
- 각각의 알파벳에 대해서, a가 처음 등장하는 위치, b가 처음 등장하는 위치, ... z가 처음 등장하는 위치를 공백으로 구분해서 출력한다.

- 만약, 어떤 알파벳이 단어에 포함되어 있지 않다면 -1을 출력한다. 단어의 첫 번째 글자는 0번째 위치이고, 두 번째 글자는 1번째 위치이다.




**첫번째 시도 -> 정답**

```python
import string

alph = string.ascii_lowercase # 알파벳 소문자 리스트 만들기 

s = input()
answer = []
for i in alph:
    if i not in s:
        answer.append('-1')
    else:
        answer.append(str(s.index(i)))  # index 함수를 이용해서 인덱스 반환

print(' '.join(answer))
```

**보다 효율적인 방법이 없을까하고 서치해본 함수 적용**

- find() : 문자열에서 적용가능하며, 없을 경우 **-1** 을 반환
    - 참고 링크 : https://ooyoung.tistory.com/68
- index() : 포함되지 않을 경우 attribute error 발생 

```python
import string

alph = string.ascii_lowercase

s = input()
answer = []
for i in alph:
    answer.append(str(s.find(i)))
    
print(' '.join(answer))
```



- 맨 위가 find 함수를 쓴 것, 아래가 index 함수를 쓴 것이다. 
- 큰 차이는 아니지만 find 함수가 더 효율적인 것으로 보인다. 

#### list.index(x[, start[, end]])

- 리스트에 있는 항목 중 값이 x 와 같은 첫 번째 것의 0부터 시작하는 인덱스를 돌려줍니다. 그런 항목이 없으면 ValueError 를 일으킵니다.

- 선택적인 인자 start 와 end 는 슬라이스 표기법처럼 해석되고, 검색을 리스트의 특별한 서브 시퀀스로 제한하는 데 사용됩니다. 돌려주는 인덱스는 start 인자가 아니라 전체 시퀀스의 시작을 기준으로 합니다.

#### find(sub[, start[, end]])
- 부분 문자열 sub 가 슬라이스 s[start:end] 내에 등장하는 가장 작은 문자열의 인덱스를 돌려줍니다. 선택적 인자 start 와 end 는 슬라이스 표기법으로 해석됩니다. sub 가 없으면 -1 을 돌려줍니다.
