# 차이를 최대로
- source : [백준 10819](https://www.acmicpc.net/problem/10819)
- 난이도 : 실버 II
- 분류 : 브루트 포스 알고리즘, 백트래킹 

## 문제
- N개의 정수로 이루어진 배열 A가 주어진다. 이때, 배열에 들어있는 정수의 순서를 적절히 바꿔서 다음 식의 최댓값을 구하는 프로그램을 작성하시오.
|A[0] - A[1]| + |A[1] - A[2]| + ... + |A[N-2] - A[N-1]|


    
## 입력
- 첫째 줄에 N (3 ≤ N ≤ 8)이 주어진다. 둘째 줄에는 배열 A에 들어있는 정수가 주어진다. 배열에 들어있는 정수는 -100보다 크거나 같고, 100보다 작거나 같다.

## 출력
- 첫째 줄에 배열에 들어있는 수의 순서를 적절히 바꿔서 얻을 수 있는 식의 최댓값을 출력한다.




**첫번째 시도**    
주어진 식을 만들어보기

```python
n = 6 
a = [20, 1, 15, 8, 4, 10]
```

```python
results = [] # 결과값 중 최댓값을 구하기 위한 리스트 

result = [] # 현재 배열의 결과값
for i in range(1, n):
    result.append(abs(a[i-1] - a[i]))
    
results.append(sum(result)) 

print(max(results)) # 최댓값 출력 
```

배열의 순서를 바꾸기(순열 만들기)
- permutation을 이용

```python
import itertools
```

```python
a_per = itertools.permutations(a)
```

```python
for i in a_per:
    print(i)
```

```python
import sys
from itertools import permutations

# n = int(sys.stdin.readline())
# a = list(map(int, sys.stdin.readline().split()))

a_per = permutations(a)

results = [] # 결과값 중 최댓값을 구하기 위한 리스트 

cnt = 0
for per in a_per:
    result = [] # 현재 배열의 결과값
    for i in range(1, n):
        result.append(abs(per[i-1] - per[i]))
    
    results.append(sum(result)) 
    
print(max(results)) # 최댓값 출력 
```
