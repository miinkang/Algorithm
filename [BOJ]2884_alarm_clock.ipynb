```python
def alarm(time):

    hour, minute = time
    
    if (minute - 45) < 0: 
        minute += 60
        
        if hour == 0:
            hour = 23
        
        else:
            hour -= 1
    
    print(hour, minute-45)
```

```python
time = [(0, 0), (10, 10), (0, 30), (23, 40)]

for t in time:
    alarm(t)
#     break
```
